# Recursion

```python
# This is a simple function to multiply two numbers without using the multiplication operator.
def multiply(a,b):
    result = 0
    for _ in range(b):
        result += a
    return result
# Example usage
print(multiply(5, 3))  # Output: 15
```

```python
# Using Recursion
def multiply(a, b):
    if b == 0:
        return 0
    else:
        result = a + multiply(a,b-1)
        return result

multiply(3,4)

def mul(a,b):
    if b == 1:
        return a
    else:
        return a + mul(a,b-1)
print(mul(6,7))  # Output: 42
```

```python
#Factorial of a number through recursion
def fact(n):
    if n == 1 or n == 0:
        return 1
    if n == 2: return 2 
    else:
        return n * fact(n-1)
    
fact(5)
```

```python
# Checking if a string is palindrome through recursion
def palindrome(s):
    if len(s) == 1:
        return 'palindrome'
    else:
        if s[0] == s[-1]:
            return palindrome(s[1:-1])
        return "not palindrome"
palindrome('malayalam')
```

```python
# Fibo
def fibo(n):
    if n == 0 or n== 1:
        return 1
    else: 
        return (fibo(n-1) + fibo(n-2))
fibo(5)
```

```python
#fibo with memo
def fibo_memo(m,d = {0:1,1:0}):
    if m in d:
        return d[m]
    else:
        d[m] = fibo_memo(m-1,d) + fibo_memo(m-2,d)
        return d[m]
print(fibo_memo(14))
```
